```python
def get_boxplot(column):
    fig, ax = plt.subplots(figsize = (14, 4))
    sns.boxplot(x=column, y='score',data=data)#.loc[data.loc[:, column].isin(data.loc[:, column].value_counts())],ax=ax)
    plt.xticks(rotation=45)
    ax.set_title('Boxplot for ' + column)
    plt.show()
```

```python
for col in spis:
    get_boxplot(col)
```

# ПРОВЕРКА ЗНАЧИМОСТИ ПРИЗНАКОВ ПРИ ПОМОЩИ КРИТЕРИЯ СТЬЮДЕНТА

```python
def get_stat_dif(column):
    cols = data.loc[:, column].value_counts()#.index[:10]
    combinations_all = list(combinations(cols, 2))
    print(column)
    for comb in combinations_all:
        if ttest_ind(data.loc[data.loc[:, column] == comb[0], 'score'], 
                        data.loc[data.loc[:, column] == comb[1], 'score']).pvalue \
            <= 0.05/len(combinations_all): # Учли поправку Бонферони
            print('Найдены статистически значимые различия для колонки', column)
            break
        #print(combinations_all)
        #print(comb[0])
        #print(data.loc[data.loc[:, column] == comb[0]])
        #print(ttest_ind(data.loc[data.loc[:, column] == comb[0], 'score'], 
         #               data.loc[data.loc[:, column] == comb[1], 'score']).pvalue,0.05/len(combinations_all))
```

```python
for col in spis:
    get_stat_dif(col)
```
