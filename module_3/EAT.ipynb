# Приводим стили кухни к нормальному виду

```python
def x_rep (x):
    x=x.replace("'","")
    x=x.replace("[","")
    x=x.replace("]","")
    return x.split(", ")
```

```python
df["Cuisine Style"]=df["Cuisine Style"].dropna().apply(x_rep)
```

```python
style=[]
for i in df["Cuisine Style"].dropna():
    style+=i
style=set(style)
```

```python
for i in style:
    df[i]=0
```

```python
for index, row in df.dropna().iterrows():
    for i in row["Cuisine Style"]:
        df.loc[index,i]=1
```

# Считаем разницу дат отзывов

```python
def date_r(x):
    x=x.split("], [")[1]
    x=x.replace("'","")
    x=x.replace("[","")
    x=x.replace("]","")
    x=x.split(", ")
    try:
        return (abs(datetime.strptime(x[0],"%m/%d/%Y").date()-datetime.strptime(x[1],"%m/%d/%Y").date()))
    except:
        return None 
```

```python
df["date_r"]=df[~df.Reviews.str.contains("[[]]")].Reviews.apply(date_r)
```

```python
df.corr()
```

```python
df["date_r"]=df["date_r"].dt.days
```

```python
df["Count_Rev"]=df["date_r"].dropna().apply(lambda x: 2)
```

```python
df["Count_Rev"]=df["Count_Rev"].fillna(0)
```

```python
df["date_r"]=df["date_r"].fillna(df["date_r"].mean())
```

# Price Range

```python
def price_to_float(x):
    try:
        if x=="$":
            return 5
        elif x=="$$ - $$$":
            return 550
        elif x=="$$$$":
            return 5000
    except:
        return x
```

```python
df["Price Range"]=df["Price Range"].apply(price_to_float)
```

```python
df["Price Range"]=df["Price Range"].fillna(500)
```

```python
data_city = pd.get_dummies(df.City)
```

```python
T=df.copy()
```

```python
T.Ranking=T.Ranking.fillna(T.Ranking.mean())
```

```python
T["Number of Reviews"]=T["Number of Reviews"].fillna(T["Number of Reviews"].median())
```

```python
T=pd.concat([T,data_city],axis=1)
```

```python
def first_model(df):
    #df["Price Range"]=df["Price Range"].dropna()#.fillna("$$ - $$$")
    X = df.drop(['Restaurant_id', 'Rating',"City","Cuisine Style","URL_TA", "ID_TA","Reviews"], axis = 1)
    y = df['Rating']
    #print(X)
    from sklearn.model_selection import train_test_split
    import numpy as np
    X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.25)
    # Х - данные с информацией о ресторанах, у - целевая переменная (рейтинги ресторанов)
   
    from sklearn.ensemble import RandomForestRegressor # инструмент для создания и обучения модели
    from sklearn import metrics # инструменты для оце
    # Создаём модель
    regr = RandomForestRegressor(n_estimators=100)

# Обучаем модель на тестовом наборе данных
    regr.fit(X_train, y_train)

# Используем обученную модель для предсказания рейтинга ресторанов в тестовой выборке.
# Предсказанные значения записываем в переменную y_pred
    y_pred = regr.predict(X_test)
    y_pred =np.round(y_pred * 2) / 2
    print('MAE:', metrics.mean_absolute_error(y_test, y_pred))
    return(y_pred)
```

```python
t=first_model(T)
```
